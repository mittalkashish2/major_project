```python
import numpy as np
import csv
import pandas as pd



def rating(interests,course_taken,rating_given):
    df=pd.read_csv(r'C:\Users\My Pc\Desktop\final_interests.csv')
    interests_superset=df[0].tolist()
    interests_superset.pop(0)
    interests_boolean=[]
    for i in range(len(interests_superset)):
        interests_boolean[i]=0
    for i in range(len(interests)):
        indexx=interests_superset.index(interests[i])
        interests_boolean[indexx]=1
    df=pd.DataFrame(interests_boolean)
    df=df.transpose()
    for i in range(len(course_taken)):
        for filename in os.listdir(r"C:\Users\My Pc\Desktop\ml_model\individual_course_files"):
            t=course_taken[i]
            if filename.endswith(t+".csv"):
                with open(r"C:/Users/My Pc/Desktop/ml_model/individual_course_files/"+t+".csv", 'a') as f:
                    new_df=df
                    ##new_df.append({'Animal':'mouse', 'Color':'black'}
                    new_df.to_csv(f,index=None,header=False) 

```
